```python
from sklearn import linear_model
from sklearn.linear_model import Lasso
from scipy.stats import norm
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt


df = pd.read_csv('/Users/dashabalashova/python_projects/1_1_5_ML_Bioinf/data/noisysine.csv')
df_tr, df_val = train_test_split(df, test_size=0.2)
df_tr = df_tr.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)

def poly_Lasso(df_tr, df_val, n, a):
    X_tr = pd.DataFrame(df_tr.x)
    X_val = pd.DataFrame(df_val.x)
    plt.scatter(df.x, df.y)
    
    for m in range(2, n+1):
        X_tr['poly_'+str(m)] = X_tr.apply(lambda s: s.x**m, axis = 1)
        X_val['poly_'+str(m)] = X_val.apply(lambda s: s.x**m, axis = 1)
    
    clf = Lasso(alpha=a, fit_intercept=True, max_iter = 1000000)
    t = np.arange(0., 40., 0.2)
    clf.fit(X_tr, df_tr.y)
    
    w = clf.coef_
    y_pl = clf.intercept_+w[0]*t
    for m in range(1, n):
        y_pl += w[m]*t**(m+1)
    plt.plot(t, y_pl, 'r')
    plt.show()
    clf.score(X_val, df_val.y)
    print('n =', n, ', alpha =', a, ', R^2 =', clf.score(X_val, df_val.y))
    print(w)
    print('not 0:', len([ x  for x in w if abs(x)>=0.00001]))

for n in range(1, 6):
    for a in [0.000000001, 0.01, 0.1, 1, 2]:
        poly_Lasso(df_tr, df_val, n, a)

```
